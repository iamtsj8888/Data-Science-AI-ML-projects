```python
data  = data.dropna()
data
```

```python
data['Estimated Employed'] = (data['Estimated Employed'] - data['Estimated Employed'].min()) / (data['Estimated Employed'].max() - data['Estimated Employed'].min()) * 19 + 1

plt.figure(figsize=(12, 10))
plt.plot(data['Estimated Employed'].iloc[:50], label='Employed', color='blue')
plt.plot(data['Estimated Unemployment Rate (%)'].iloc[:50], label='Unemployed', color='red')
plt.legend()
plt.show()
```

```python
correlation = data['Estimated Employed'].corr(data['Estimated Unemployment Rate (%)'])
print("Correlation between Employed and Unemployment Rate:", correlation)
```
